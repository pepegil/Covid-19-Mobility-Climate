```python
def formato_prec(row):
    if row.prec== 'nan':
        precN =0.0
    else :
        litros = float(row.prec[:-2])
        decima = float(row.prec[-1:])
        precN = (litros) + (decima)/10.0
    #print(precN)
    return precN 
```
